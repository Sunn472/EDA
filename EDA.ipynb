```python
'''
Q11. Is there any seasonal trend in car sales based on the date of sale
'''
df1['Date'] = pd.to_datetime(df1['Date'], format='%m/%d/%Y')

#Extract month and year from Date
df1['Month'] = df1['Date'].dt.month
df1['Year'] = df1['Date'].dt.year

#Define seasons based on months
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

df1['Season'] = df1['Month'].apply(get_season)

# Count sales by month
monthly_sales = df1['Month'].value_counts().sort_index()
monthly_sales.index = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
print("Number of Cars Sold by Month:")
print(monthly_sales)

#Count sales by season
seasonal_sales = df1['Season'].value_counts().reindex(['Winter', 'Spring', 'Summer', 'Fall'])
print("\nNumber of Cars Sold by Season:")
print(seasonal_sales)

#Visualize sales by month
plt.figure(figsize=(12, 6))
sns.barplot(x=monthly_sales.index, y=monthly_sales.values)
plt.title('Car Sales by Month')
plt.xlabel('Month')
plt.ylabel('Number of Cars Sold')
plt.tight_layout()
plt.show()

# Visualize sales by season
plt.figure(figsize=(10, 6))
sns.barplot(x=seasonal_sales.index, y=seasonal_sales.values)
plt.title('Car Sales by Season')
plt.xlabel('Season')
plt.ylabel('Number of Cars Sold')
plt.tight_layout()
plt.show()

#Visualize monthly sales trend over years (if multiple years exist)
yearly_monthly_sales = df1.groupby(['Year', 'Month']).size().unstack().fillna(0)
plt.figure(figsize=(14, 8))
sns.heatmap(yearly_monthly_sales, cmap='Blues', annot=True, fmt='.0f')
plt.title('Car Sales by Month and Year')
plt.xlabel('Month')
plt.ylabel('Year')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.tight_layout()
plt.show()
```

```python
'''
Q1.What is the average rating for each product category
'''

df2['rating'] = df2['rating'].str.replace('|', '')

df2['rating'] = pd.to_numeric(df2['rating'], errors='coerce')

avg_rating=df2.groupby('category').rating.mean()
print(f'the average rating for each product category: {avg_rating}')
```

```python
'''
Q2. What are the top rating_count products by category
'''
top_rating_count=df2.groupby('category').rating.max()
print(f'the top rating_count products by category: {top_rating_count}')
```

```python
'''
Q3. What is the distribution of discounted prices vs. actual prices
'''
df2['discounted_price']=df2['discounted_price'].str.replace('â‚¹','').str.replace(',','')
df2['actual_price']=df2['actual_price'].str.replace('â‚¹','').str.replace(',','')
df2['discounted_price']=df2['discounted_price'].astype(float)
df2['actual_price']=df2['actual_price'].astype(float)

sns.scatterplot(x=df2['discounted_price'],y=df2['actual_price'])
plt.title('Distribution of discounted prices vs. actual prices')
plt.xlabel('Discounted Price')
plt.ylabel('Actual Price')
plt.show()
```

```python
'''
Q4. How does the average discount percentage vary across categories
'''

df2['discount_percentage']=df2.discount_percentage.str.replace('%','')
df2['discount_percentage']=df2['discount_percentage'].astype(float)


# Extract top-level category (before the first '|')
df2['main_category'] = df2['category'].str.split('|').str[0]

# Calculate average discount by main category
avg_discount_by_category = df2.groupby('main_category')['discount_percentage'].mean().sort_values(ascending=False)

# Visualization
plt.figure(figsize=(12, 6))
avg_discount_by_category.plot(kind='bar', color='skyblue')
plt.title('Average Discount Percentage by Main Category')
plt.xlabel('Category')
plt.ylabel('Average Discount Percentage')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Display the numerical results
print("Average Discount Percentage by Category:")
print(avg_discount_by_category.round(2))



```

```python
'''
Q5. What are the most popular product names
'''

df2['rating_count'] = df2['rating_count'].str.replace(',', '')
df2['rating_count']=df2['rating_count'].astype(float)

product_popularity = df2.groupby('product_name').agg({
    'rating_count': 'sum',
    'rating': 'mean'}).sort_values('rating_count', ascending=False).head(1)


# Display the top products
print("Top Most Popular Products:")
print(product_popularity[['rating_count', 'rating']].head(1))
```

```python
'''
Q6.What are the most popular product keywords
'''
from collections import Counter
import re

def extract_keywords(text):
    # Remove special characters and split into words
    words = re.findall(r'\w+', text.lower())
    # Remove common stop words
    stop_words = {'and', 'for', 'the', 'with', 'this', 'that', 'your', 'you', 'are', 'from', 'has', 'have', 'was', 'were', 'its', 'it', 'in', 'on', 'at', 'to', 'of', 'a', 'an', 'is', 'as', 'be', 'by'}
    return [word for word in words if word not in stop_words and len(word) > 2]

# Combine product names and about_product for keyword analysis
all_text = ' '.join(df2['product_name'].fillna('')) + ' ' + ' '.join(df2['about_product'].fillna(''))
keywords = extract_keywords(all_text)

# Get top keywords
keyword_counts = Counter(keywords)
top_keywords = keyword_counts.most_common(1)

print("\nTop Product Keywords:")
for word, count in top_keywords:
    print(f"{word}: {count}")

```

```python
'''
Q7. What are the most popular product reviews
'''
from textblob import TextBlob


df2['review_content'] = df2['review_content'].str.lower().str.replace('[^\w\s]', '')


df2['review_length'] = df2['review_content'].str.split().str.len()


def get_common_phrases(text_series, n=3, top=20):
    phrases = []
    for text in text_series.dropna():
        words = text.split()
        phrases.extend([' '.join(words[i:i+n]) for i in range(len(words)-n+1)])
    return Counter(phrases).most_common(top)

common_phrases = get_common_phrases(df2['review_content'])

print("Most Common Review Phrases:")
for phrase, count in common_phrases:
    print(f"{phrase}: {count}")
```

```python
'''
Q8. What is the correlation between discounted_price and rating
'''
corr=df2['discounted_price'].corr(df2['rating'])
print(f'The correlation between discounted_price and rating: {corr}')

sns.heatmap(df2[['discounted_price','rating']].corr(),annot=True,cmap='coolwarm',fmt='.2f')
plt.show()
```

```python
'''
Q9. What are the Top 5 categories based on the highest rating
'''

def get_primary_category(category_str):
    if pd.isna(category_str):
        return None
    primary = category_str.split('|')[0]
    return primary.strip()

df2['primary_category'] = df2['category'].apply(get_primary_category)


category_ratings = df2.groupby('primary_category')['rating'].mean().sort_values(ascending=False)

top_5_categories = category_ratings.head(5)

print("Top 5 Categories by Average Rating:")
print(top_5_categories)
```

```python
'''
Q2. What is the distribution of popularity among the tracks in the dataset?
Visualize it using a histogram
'''
df3_copy=df3.copy()
print("Popularity statistics:")
print(df3_copy['Popularity'].describe())

# Plot histogram of popularity
plt.figure(figsize=(12, 6))
sns.histplot(df3_copy['Popularity'], bins=30, kde=True, color='#1DB954')
plt.title('Distribution of Track Popularity', fontsize=15)
plt.xlabel('Popularity Score (0-100)', fontsize=12)
plt.ylabel('Number of Tracks', fontsize=12)
plt.axvline(df3_copy['Popularity'].mean(), color='red', linestyle='--',
            label=f'Mean: {df3_copy["Popularity"].mean():.1f}')
plt.axvline(df3_copy['Popularity'].median(), color='green', linestyle='--',
            label=f'Median: {df3_copy["Popularity"].median():.1f}')
plt.legend()
plt.show()
```

```python
'''
Q3. Is there any relationship between the popularity and the duration of tracks?
Explore this using a scatter plot
'''


print("Summary Statistics:")
print(df3_copy[['Popularity', 'Duration (ms)']].describe())

# Correlation between popularity and duration
correlation = df3_copy['Popularity'].corr(df3_copy['Duration (ms)'])
print(f"\nCorrelation between Popularity and Duration: {correlation:.3f}")

# Create scatter plot
plt.figure(figsize=(12, 7))
sns.scatterplot(data=df3_copy, x='Duration (ms)', y='Popularity', alpha=0.6, color='royalblue')

# Add title and labels
plt.title('Relationship Between Track Popularity and Duration', fontsize=14)
plt.xlabel('Duration (miliseconds)', fontsize=12)
plt.ylabel('Popularity Score', fontsize=12)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
'''
Q4. Which artist has the highest number of tracks in the dataset? Display the
count of tracks for each artist using a countplot
'''


artist_counts = df3_copy['Artist'].value_counts().reset_index()
artist_counts.columns = ['Artist', 'Track Count']

# Get top 10 artists by track count
top_artists = artist_counts.head(10)

print("Top 10 Artists by Number of Tracks:")
print(top_artists.to_string(index=False))

# Create countplot for top artists
plt.figure(figsize=(12, 6))
sns.barplot(data=top_artists, x='Track Count', y='Artist', palette='viridis')
plt.title('Top 10 Artists by Number of Tracks in Dataset', fontsize=14)
plt.xlabel('Number of Tracks', fontsize=12)
plt.ylabel('Artist', fontsize=12)
plt.grid(axis='x', alpha=0.3)


for i, count in enumerate(top_artists['Track Count']):
    plt.text(count+0.5, i, str(count), va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
'''
Q5.What are the top 5 least popular tracks in the dataset? Provide the artist
name and track name for each
'''
least_popular = df3.sort_values('Popularity').head(5)[['Artist', 'Track Name', 'Popularity']]

print("Top 5 Least Popular Tracks:")
print(least_popular.to_string(index=False))
```

```python
'''
Q6. Among the top 5 most popular artists, which artist has the highest
popularity on average? Calculate and display the average popularity for each artist
'''

top_5_artists = df3_copy['Artist'].value_counts().head(5).index.tolist()

top_artists_data = df3_copy[df3_copy['Artist'].isin(top_5_artists)]

avg_popularity = top_artists_data.groupby('Artist')['Popularity'].mean().sort_values(ascending=False).reset_index()

print("Average Popularity for Top 5 Artists (by track count):")
print(avg_popularity.to_string(index=False, float_format="%.2f"))
```

```python
'''
Q7. For the top 5 most popular artists, what are their most popular tracks? List
the track name for each artist
'''
top_5_artists = df3_copy['Artist'].value_counts().head(5).index.tolist()

most_popular_tracks = df3_copy[df3_copy['Artist'].isin(top_5_artists)] \
    .sort_values('Popularity', ascending=False) \
    .drop_duplicates('Artist') \
    [['Artist', 'Track Name', 'Popularity']] \
    .sort_values('Popularity', ascending=False)

print("Most Popular Tracks for Top 5 Artists:")
print(most_popular_tracks.to_string(index=False))
```

```python
'''
Q8. Visualize relationships between multiple numerical variables simultaneously
using a pair plot
'''

top_artists = df3_copy['Artist'].value_counts().head(5).index
df_top = df3_copy[df3_copy['Artist'].isin(top_artists)]

# Create pair plot with hue by artist
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")
pair_grid = sns.pairplot(data=df_top,
                        vars=['Popularity', 'Duration (ms)'],
                        hue='Artist',
                        palette='viridis',
                        height=4,
                        aspect=1.2,
                        plot_kws={'alpha': 0.7, 's': 80})

pair_grid.fig.suptitle('Relationships Between Popularity and Duration by Top Artists', y=1.02)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(data=df_top, x='Artist', y='Popularity', palette='viridis')
plt.title('Distribution of Popularity Scores by Top Artist')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
'''
Q9.Does the duration of tracks vary significantly across different artists?
Explore this visually using a box plot or violin plot
'''

# Get top 10 artists by track count
top_artists = df3_copy['Artist'].value_counts().head(10).index
df_top = df3_copy[df3_copy['Artist'].isin(top_artists)]

# Set style
sns.set(style="whitegrid", palette="pastel")
plt.figure(figsize=(14, 8))

# Box plot
plt.subplot(1, 2, 1)
sns.boxplot(data=df_top, x='Duration (ms)', y='Artist')
plt.title('Distribution of Track Durations by Artist (Box Plot)')
plt.xlabel('Duration (minutes)')
plt.ylabel('Artist')

# Violin plot
plt.subplot(1, 2, 2)
sns.violinplot(data=df_top, x='Duration (ms)', y='Artist')
plt.title('Distribution of Track Durations by Artist (Violin Plot)')
plt.xlabel('Duration (minutes)')
plt.ylabel('')

plt.tight_layout()
plt.show()

# Statistical summary
duration_stats = df_top.groupby('Artist')['Duration (ms)'].describe()[['count', 'mean', 'std', 'min', '50%', 'max']]
print("\nDuration Statistics by Artist (minutes):")
print(duration_stats.sort_values('mean', ascending=False).to_string(float_format="%.2f"))
```

```python
'''
Q10. How does the distribution of track popularity vary for different artists?
Visualize this using a swarm plot or a violin plot.
'''
# Filter for artists with at least 5 tracks
artist_counts = df3_copy['Artist'].value_counts()
top_artists = artist_counts[artist_counts >= 5].index.tolist()
filtered_df = df3_copy[df3_copy['Artist'].isin(top_artists)]

# Create the plot
plt.figure(figsize=(12, 8))
sns.violinplot(x='Artist', y='Popularity', data=filtered_df, inner='quartile')

# Customize the plot
plt.title('Distribution of Track Popularity by Artist', fontsize=14)
plt.xlabel('Artist', fontsize=12)
plt.ylabel('Popularity Score', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
